```python
import math

class Circulo:
    def __init__(self, radio):
        self.radio = radio
        self.perimetro = 2 * math.pi * self.radio
        self.area = math.pi * self.radio ** 2
```

```python
#Uso de la clase
C = Circulo(3)
print(C.perimetro, C.area)
```

```python
class Punto3D:
    """
    Clase Punto3D que inicializa las coordenadas x, y y z.
    """
    def __init__(self, x=0.0, y=0.0, z=0.0):
        # Inicialización de los argumentos
        self.x = x
        self.y = y
        self.z = z

    def __str__(self):
        # Para que imprima el valor y no la dirección
        return f'({self.x}, {self.y}, {self.z})'

    def distancia(self, otro_punto):
        distx = otro_punto.x - self.x
        disty = otro_punto.y - self.y
        distz = otro_punto.z - self.z
        return (distx**2 + disty**2 + distz**2)**0.5
```

Como la función distancia está definida dentro de la clase es necesario crear instancias de la clase Punto3D para ambos puntos (diseñada para trabajar con objetos de esta clase).

```python
# Ejemplo
A = Punto3D(4, -3, 2)
B = Punto3D(1, 2, -5)

# Accediendo a las coordenadas y calculando la distancia
A.x, A.y, A.z, B.x, B.y, B.z, A.distancia(B)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Se realizará marcando puntos en dos circunferencias y luego uniéndolos

r_externo = 6      # Radio del círculo exterior
r_interno = 3      # Radio del círculo interior

# Definir ángulos

n_puntas = 5 #mayor o igual a 4 para que "parezca" una estrella (subjetivo)
ángulo_entre_puntas = 2 * np.pi / n_puntas

# Modificación para que la estrella quede apuntando hacia arriba
ángulos = np.arange(np.pi / 2, 5 * np.pi / 2, ángulo_entre_puntas / 2)

# Coordenadas (se irá guardando las coordenadas de los puntos en listas)
x_coords = []
y_coords = []

# Calcular puntos alternando entre los radios externo e interno (alternando entre par e impar)
for i, ángulo in enumerate(ángulos):
    if i % 2 == 0:
        # Puntos exteriores
        x = r_externo * np.cos(ángulo)
        y = r_externo * np.sin(ángulo)
    else:
        # Puntos interiores
        x = r_interno * np.cos(ángulo)
        y = r_interno * np.sin(ángulo)
    x_coords.append(x)
    y_coords.append(y)


# Crear la figura y el eje
fig, ax = plt.subplots()

# Dibujar la estrella y modificación de colores
ax.fill(x_coords, y_coords, color='yellow', edgecolor='orange')

# Establecer aspecto igual para que la estrella no se distorsione, esto sirve para que las proporciones estén bien.
ax.set_aspect('equal')

# Ocultar ejes
ax.axis('off')

# Mostrar la figura
plt.show()
```

```python
import pandas as pd

# Datos
datos = {
    'nombre': ['José Luis Martínez Izquierdo', 'Rosa Díaz Díaz', 'Javier García Sánchez',
               'Carmen López Pinzón', 'Marisa López Collado'],
    'edad': [18, 32, 24, 35, 46],
    'sexo': ['H', 'M', 'H', 'M', 'M'],
    'peso': [85.0, 65.0, None, 65.0, 51.0],
    'altura': [1.79, 1.73, 1.81, 1.70, 1.58],
    'colesterol': [182.0, 232.0, 191.0, 200.0, 148.0]
}

df = pd.DataFrame(datos)

df.head()
```

```python
#5(b). Solo mujeres y calcule el promedio de su colesterol

mujeres = df[df['sexo'] == 'M']
promedio_colesterol_mujeres = mujeres['colesterol'].mean()

print("El promedio del colesterol de las mujeres es:", promedio_colesterol_mujeres)
mujeres.head()

```
