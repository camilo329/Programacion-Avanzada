**Obtener una lista o matriz de todos los estados en el conjunto de datos, sin repetición, ordenados de la A a la Z**

```python
lista_estados = sorted(df['ST'].unique())
print(lista_estados)
```

**¿Cuáles son los 5 estados con más bancos fallidos?**

```python
# Cantidad de bancos fallidos por estado
bancos_por_estado = df['ST'].value_counts()

# 5 estados con más bancos fallidos
top_5_estados = bancos_por_estado.head(5)

#Resultados
print(top_5_estados)
```

**¿Cuáles son las 5 principales instituciones adquirentes?**

```python
# Contar la cantidad de veces que cada institución adquirente aparece
instituciones_adquirentes = df['Acquiring Institution'].value_counts()

# Obtener las 5 principales instituciones adquirentes
top_5_instituciones = instituciones_adquirentes.head(5)

# Mostrar los resultados
print(top_5_instituciones)
```

**¿Cuántos bancos ha adquirido el State Bank of Texas? ¿Cuántos de ellos estaban realmente en Texas?**

```python
# Filtrar los bancos adquiridos por 'State Bank of Texas'
bancos_state_bank_of_texas = df[df['Acquiring Institution'] == 'State Bank of Texas']

# Contar cuántos bancos ha adquirido el 'State Bank of Texas'
total_bancos_adquiridos = bancos_state_bank_of_texas.shape[0]

# Contar cuántos de esos bancos están en Texas (TX)
bancos_en_texas = bancos_state_bank_of_texas[bancos_state_bank_of_texas['ST'] == 'TX'].shape[0]

# Mostrar los resultados
print(f"El State Bank of Texas ha adquirido {total_bancos_adquiridos} bancos.")
print(f"De esos, {bancos_en_texas} estaban realmente en Texas.")
```

**¿Cuál es la ciudad más común en California donde un banco quiebra?**

```python
# Filtrar por bancos en California
bancos_california = df[df['ST'] == 'CA']

# Encontrar la ciudad más común
ciudad_comun = bancos_california['City'].mode()[0]

print(f"La ciudad más común en California donde un banco quiebra es: {ciudad_comun}")
```

**¿Cuántos nombres de bancos comienzan con la letra 's'?**

```python
# Filtrar bancos cuyos nombres comiencen con 'S'
bancos_con_s = df[df['Bank Name'].str.startswith('S')]

# Contar cuántos bancos cumplen esta condición
cantidad_bancos_con_s = bancos_con_s.shape[0]

print(f"Hay {cantidad_bancos_con_s} bancos cuyos nombres comienzan con la letra 'S'.")
```

**¿Cuántos bancos consisten en dos palabras? Ej. "First Bank" , "Bank Georgia"**

```python
# Contar los bancos cuyos nombres tienen exactamente dos palabras
bancos_dos_palabras =df['Bank Name'].apply(lambda x: len(x.split()) == 2)

# Contar cuántos bancos cumplen esta condición
cantidad_bancos_dos_palabras = bancos_dos_palabras.sum()

print(f"Hay {cantidad_bancos_dos_palabras} bancos cuyos nombres consisten en dos palabras.")
```

**¿Cuántos bancos quebrados no tienen la palabra "Banco" en su nombre?**

```python
# Filtrar bancos cuyos nombres no contienen la palabra "Bank"
bancos_sin_bank = df[~df['Bank Name'].str.contains('Bank', case=False)]

# Contar cuántos bancos cumplen esta condición
cantidad_bancos_sin_bank = bancos_sin_bank.shape[0]

print(f"Hay {cantidad_bancos_sin_bank} bancos quebrados que no tienen la palabra 'Bank' en su nombre.")
```

**¿Cuántos bancos cerraron en el año 2008?**

```python
# Especificar el formato de fecha, por ejemplo 'MM/DD/YYYY'
df['Closing Date'] = pd.to_datetime(df['Closing Date'], format='%m/%d/%Y', errors='coerce')

# Filtrar por bancos que cerraron en 2008
df_2008 = df[df['Closing Date'].dt.year == 2008]

# Cantidad de bancos que cerraron en 2008
cantidad_bancos_2008 = df_2008.shape[0]

print(f"En el año 2008 cerraron {cantidad_bancos_2008} bancos.")

#El parámetro errors='coerce' en pd.to_datetime convierte los valores no válidos o mal formateados en NaT (valores nulos) durante la conversión a formato datetime, evitando que el proceso falle.
```
