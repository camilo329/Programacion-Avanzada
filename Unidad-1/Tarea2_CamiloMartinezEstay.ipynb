Clase Punto 2D

```python
class Punto2D:
    """
    Clase Punto2D que inicializa las coordenadas x e y.
    """
    def __init__(self, x=0.0, y=0.0):
      #Inicialización de los argumentos
        self.x = x
        self.y = y

    def __str__(self):
      #Para que imprima el valor y no la dirección
        return f'({self.x}, {self.y})'

    def distancia(self, otro_punto):
        distx = otro_punto.x - self.x
        disty = otro_punto.y - self.y
        return (distx**2 + disty**2)**0.5
```

Como la función distancia está definida dentro de la clase es necesario crear instancias de la clase Punto2D para ambos puntos (diseñada para trabajar con objetos de esta clase).
