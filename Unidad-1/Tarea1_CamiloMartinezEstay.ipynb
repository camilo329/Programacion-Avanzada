Gráfica 1: Elipse centrada en el origen

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parámetros de la elipse
a = 3
b = 5

# Validación de los parámetros (la función raise se usa para indicar que se ha producido un error o una condición excepcional)
if a <= 0 or b <= 0:
    raise ValueError("Los valores de a y b deben ser mayores que cero.")


# Valores de x
x = np.linspace(-a, a, 200)


# Cálculo de elipse superior e inferior
elipse_sup = b * np.sqrt(1 - (x**2 / a**2))
elipse_inf = -elipse_sup

# Graficar
plt.plot(x, elipse_sup, color='green', label=f'$\\frac{{x^2}}{{{a**2}}} + \\frac{{y^2}}{{{b**2}}} = 1$')
plt.plot(x, elipse_inf, color='green')

# Contenido de la gráfica
plt.title(f'Elipse centrada en (0,0) con semiejes de valores {a} y {b}')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```

Gráfica 2: función $f(x, y) = 9x^2 - 5y^2$ en **3D**.

```python
# Para visualizaciones interactivas:
import plotly.graph_objects as go
```

```python
# Dominio de la función
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x,y)

# Definición de función a graficar (representación como matriz)
Z = 9 * X**2 - 5 * Y**2

# Gráfica
fig = go.Figure(data=[go.Surface(x=X, y=Y,z=Z,colorscale='sunset',opacity=0.7)])
fig.update_layout(scene = dict(
    xaxis_title='Eje x',  # Etiqueta Eje x
    yaxis_title='Eje y',  # Etiqueta Eje y
    zaxis_title='Eje z'), # Etiqueta Eje z
    title='$\hbox{Grafica de } f(x, y) = 9x^2 - 5y^2$ en **3D**.$',
    width=800,
    margin=dict(r=0, b=20, l=30, t=40)
)
fig.show()
```
